# Tema 1: Aproximaciones númericas y errores 

Un método numérico en un proceso matemático iterativo cuyo objetivo es encontrar la aproximación a una solución específica con un cierto error previamente determinado.

A  diferencia de las técnicas propias de la matemática analítica, los
métodos numéricos requieren de una aproximación a la solución
real del problema, misma que es corregida a través de la repetición de un cierto proceso que debe arrojar soluciones cada vez más
cercanas al valor real. Cada corrección de un valor inicial se conoce
como iteración. El proceso es controlado por medio de la medición
de una cantidad de error predefinido entre dos aproximaciones
sucesivas.

### Aproximación numérica y errores 
Una aproximación es un valor cercano a uno considerado como real o verdadero. Esta cercanía o diferencia se conoce como error.

### Exactitud y precisión 
La exactitud se refiere a que tan cercano está el valor calculado o medido del valor verdadero. La precisión se refiere a que tan cercanos se encuentran, unos de otros, diversos valores calculados o medidos.

### Errores
Los errores son las diferencias entre los resultados experimentales y los emanados propiamente del modelo matemático. A las diferencias cuantitativas entre en los métodos se les denomina errores.

* Error absoluto  
El error absoluto es definido como $e_a$, se obtiene como el valor absoluto de la diferencia entre el valor exacto $x$ y la aproximación $x_a$. Esto es: $$e_a=|x-x_a|$$  La diferencia se toma en valor absoluto, debido a que no interesa si la aproximación es mayor o menor que el valor exacto.  


* Error relativo  
El error relativo se define como el error absoluto del cociente de la diferencia $x-x_a$ entre el valor exacto. Esto es: $$e_r=|\frac{x-x_a}{x}|$$  


### Clasificación de los errores 
 Los errores pueden ser clasificados de acuerdo a su origen en los siguientes tres tipos:
 * Errores inherentes: Estos son errores por la forma de lectura del aparato utilizado o por errores humanos.
 * Errores por truncamiento: se presentan al aproximar funciones analíticas por medio de algunos términos de una serie infinita.
 * Errores por redondeo: es consecuencia directa de redondear a un determinado número de cifras decimales las cantidades que se operan en un proceso de solución.

### Convergencia 
Se dice que una sucesión de números $x_1,x_2,x_3,...,x_n$ converge a un valor x, si para todo $\epsilon>0$ existe un entero m, tal que para todo entero n>m se cumple que: $$|x-x_n|<\epsilon$$  
Para un método de aproximaciones sucesivas no es posible aplicar la definición anterior, pero se puede determinar si el método converge, la diferencia en valor absoluto de las dos últimas aproximaciones es menor que la diferencia en valor absoluto entre la penúltima y la antepenúltima aproximación, es decir... $$|x_n-x_{n-1}| < |x_{n-1}-x_{n-2}|$$  
Si la convergencia se cumple en los primeros valores de n se puede esperar que la convergencia continúe. También se puede demostrar que si un método es convergente, la diferencia en valor absoluto entre las últimas dos aproximaciones es mayor o igual que el error absoluto; es decir, que esta diferencia es una cota del error absoluto, con lo cual se puede escribir: $$|x_n - x_{n-1}| \geq e_a $$

## Aproximación de funciones por medio de polinomios 

Para encontrar la función polinómica $P$ que aproxime a otra función $f$, se empieza por elegir un número c en el dominio de $f$ en el que $P$ y $f$ tengan el mismo valor: $P(c) = f(c)$ esto es que ambas funciones pasan por el mismo punto $(c,f(c))$.

Se dice entonces que la aproximación polinómica se expande alrededor de $c$ está centrada en $c$. Geométricamente, el requisito de que $P(c) = f(c)$ significa que la gráfica de $P$ debe pasar por el punto $(c,f(c))$. Hay muchos polinomios cuya gráfica pasa por ese punto; la tarea es encontrar un polinomio cuya gráfica se parezca a la función $f$ en la cercanía de ese punto. Una manera de hacer esto es imponer el requisito adicional de que la pendiente de la función polinómica sea la misma que la pendiente de la gráfica de $f$ en el punto $(c,f(c))$. Dicho de otro modo: $$P'(c)=f'(c)$$

### Polinomio de Taylor 
Si $f$ tiene $n$ derivada en $c$ entonces el polinomio: $$P_n(x) = F(c)+f'(c)(x-c)+\frac{f''(c)}{2!}(x-c)^2+...+\frac{f^n(c)}{n!}(x-c)^n$$  
Se llama polinomio de Taylor de grado n para la función $f$ en el punto c.

### Polinomio de MacLaurin 
Si $c=0$, entonces: $$P_n(x) = F(0)+f'(0)+\frac{f''(0)}{2!}x^2+...+\frac{f^n(0)}{n!}x^n$$

 ## Método de Bisección
Cuando se observa que $f(x)$ cambia de signo en ambos lados de la raíz en general $f(x)$ es real y continua en el intervalo que va desde $X_o$ hasta $X_n$ y $f(x_i)$ y $f(X_n)$ tienen signos opuestos es decir: $$f(x_i)f(x_u) < 0$$  entonces hay al menos una raíz entre $x_i$ y $x_u$
El método de bisección, es un tipo de búsqueda incremental en el que el intervalo se divide siempre a la mitad. Si la función cambia de signo sobre un intervalo, se evalúa el valor de la función en el punto medio. La posición de la raíz se determina situándola en el punto medio del subintervalo, dentro del cual ocurre un cambio de signo. El proceso se repite hasta obtener una mejor aproximación. 

**<span style="color:purple">Algoritmo básico de método de bisección:</span>**
<br>
Paso 1:
> Elija valores iniciales inferior y superior ($x_i, x_u$) que encierren a la raíz de tal forma que la funcion cambie de signo en el intervalo. Esto se verifica comprobando que $f(x_i)f(x_u)>0$

Paso 2: Una aproximación de la raíz x, se determina mediante: 
<div class="alert alert-block alert-warning">
<b>$$x_r 
= \frac{x_l+x_u}{2} $$  
</div>
Paso 3: Realice las siguientes evaluaciones para determinar en qué subintervalo está la raíz:  
<br>  a) Si $f(x_i)f(x_u)<0$, entonces la raíz se encuentra dentro del subintervalo inferior o izquierdo. Por lo tanto, haga $x_u =x_r$, y vuelva al paso 2.  
<br>   b) Si $f(x_i)f(x_u)>0$, entonces la raíz se encuentra dentro del subintervalo superior o derecho. Por lo tanto, haga $x_i =x_r$, y vuelva al paso 2.  
<br>    c) Si $f(x_i)f(x_u)=0$, La raíz es igual a $x_r$ termina el cálculo. 
    
    

### Código para determinar la raíz real positiva por método de bisección
**<span style="color:green">Ejemplo:</span>** Determinar la raíz real positiva de la siguiente ecuación, con dos cifras decimales exactas. $$x^3-x-1=0$$

```python
import  math 
def f(x):
    return x**3-x-1 #Se cambia la funcion de x
def biseccion(xl,xu,tol):
    xold=100
    xr=101
    i=0
    if(f(xl)*f(xu)>0):
        print("no existe una raíz real en el intervalo, cambia los limites")
    else:
        while(abs(xr-xold)>tol):
            i=i+1
            xold=xr
            xr=(xl+xu)/2
            if(f(xl)*f(xr)<0):
                xu=xr
            elif(f(xl)*f(xr)>0):
                xl=xr
            else:
                print("la raiz es:", xr)
        #print(round(xr,6))
    print("La aproximación de la raíz R en la iteracion",i,"es:",round(xr,6))
```

## Método de la falsa posición  (interpolación lineal)
La falsa posición es una alternativa basada en una visualización gráfica.  
Consiste en unir $F(x_i)$ y $F(x_u)$ con una línea recta. La interseccion de esta línea con el eje de las x representa una mejor aproximación a la raíz. El hecho de que se remplace la curva por una línea recta da una "falsa posición" de la raíz.

 
 $$Generado \ por \ Paint \ y \ Power \ Point$$
 Usando triángulos semejantes, la intersección de la línea recta con el eje de las x se estima mediante $$\frac{f(x_i}{x_r-x_i} = \frac{f(x_u}{x_r-x_u}$$
 en la cual se despeja $x_r$ $$x_r = x_u - \frac{f(x_u)(x_i-x_u)}{f(x_i)-f(x_u)}$$
 Esta es la fórmula de la falsa posición.

Paso 2: Una aproximación de la raíz x, se determina mediante: 
<div class="alert alert-block alert-warning">
<b>$$x_r = x_u - \frac{f(x_u)(x_i-x_u)}{f(x_i)-f(x_u)}$$
</div>

### Código para determinar la raíz real positiva por metodo de falsa posición

## Métodos abiertos 
Los métodos abiertos descritos a continuación, a diferencia de los métodos cerrados, se basan en fórmulas que requieren únicamente de un solo valor de inicio $x_0$ que empiece con un par de ellos, pero que no necesariamente encierran a la raíz. Éstos, algunas veces divergen o se alejan de la raíz verdadera a medida que se avanza en el cálculo, pero cuando los métodos abiertos convergen, por lo general lo hacen mucho más rapido que los métodos cerrados.

## Método de punto fijo 
El método de punto fijo se considera como el más simple en sus principios y en el se pueden apreciar claramente todas las características de un método de aproximaciones sucesivas.

Sea $f(x)= 0$ una ecuación algebraica o trascendente cualquiera. Se
suma $x$ en ambos miembros y se obtiene:
$$f(x) + x = x$$ donde el miembro izquierdo es otra función de $x$ que se define como: $$g(x) = f(x) + x$$ por lo que... $$x = g(x)$$ Observen que cualquier ecuación puede representarse en esta forma,
siguiendo el procedimiento anterior. 
Si $x = a$ es una raíz de la ecuación, entonces $$f(a) = 0$$ y por ende, $$a=g(a)$$
El método de aproximaciones sucesivas consiste en sustituir un valor inicial $(x_0)$ apropiado (cercano a la raíz) en el segundo miembro de la ecuación anterior, es decir: $$x_0=g(x_0)$$
Sin embargo, esto será difícil que ocurra; seguramente el valor inicial proporcionado $x_0$ será sólo un valor cercano a la raíz. Entonces, en el caso más general: $$x_0 ≠ g(x_0); x_1 = g(x_0)$$
donde $x_1$ es la nueva aproximación de la raíz $a$. Se sustituye $x_1$ en el segundo miembro de la ecuación $x = g(x)$ y se obtiene:
$$x_2 = g(x_1)$$ Al proceder reiteradamente en esta forma se induce que la n-ésima aproximación es:
<span style="color:blue">$$x_n = g(x_{n-1});$$</span>; $ n = 1,2,3,...$

### Código para determinar la raíz de una ecuación por método de punto fijo 
**<span style="color:green">Ejemplo:</span>** Haciendo uso del método de punto fijo, localizar la raíz de la siguiente
ecuación. $$f(x)=e^{-x}-x$$

```python
#Ecuación de recurrencia xn=e^-xn-1
import math 

def funcion(x):
    return math.exp(-x)

x_l = 0
x_u = 1
x_ap = (x_l + x_u) / 2
x_ant = 1000
it = 1
while abs(x_ap- x_ant) > 0.0001:
    x_ant = x_ap
    x_ap = funcion(x_ap)
    print("Iteración:", it)
    it += 1
    print(round(x_ap,8))
```

## Método de Newton - Raphson
Mejores características de eficiencia, casi siempre converge a la solución y lo hace en un número reducido de iteraciones.
Parte de una primera aproximación $x_i$ y mediante la aplicación de una fórmula de recurrencia se acerca a la raíz buscada, de tal manera que la nueva aproximación $x_{i+1}$ se localiza en la intersección de la tangente a la curva de la función en el punto $x_n$ y en el eje de las abscisas.



 
$$Generado \ por \ Paint \ y \ Power \ Point$$
Para deducir la fórmula de recurrencia, en la figura se puede observar que:
 $$tan(\theta) = f'(x_i)$$
Además, por definición de la tangente:
$$tan(\theta)=\frac{f(x_i)}{x_i-x{i+1}}$$
Igualando las dos expresiones anteriores, tenemos:
$$f'(x_i)=\frac{f(x_i)}{x_i-x{i+1}}$$
Por lo tanto, la fórmula de recurrencia es:  

<center> <span style="color:blue">$$x_{i+1}=x_i-\frac{f(x_i)}{f'(x_i)};$$</span> $$i =0,1,2,3,...$$ 

Esta ecuación es la fórmula de recurrencia del método de Newton-Raphson,
conocido como el el método de las tangentes.

### Código para determinar la raíz de una ecuación por método de Newton Raphson
**<span style="color:green">Ejemplo:</span>** Obtener la raíz real posi:va de la siguiente ecuación trascendente, con
tres cifras decimales exactas.
ecuación. $$2cos(x)-e^x=0$$

```python
import math 
def f(x):
    return 2*math.cos(x)-math.exp(x)
def df(x):
    return -2*math.sin(x)-math.exp(x)
def nRaphson(x0,tol):
    xa=0
    x=x0
    i=0
    while abs (x-xa)>tol:
        i=i+1
        xa=x
        x=x-(f(x)/df(x))
        print("Iteración:", i)
        print(round(x,8))
```

## Versión Doolittle

- L: Matriz triangular inferior con elementos unitarios en la diagonal principal.
- U: Matriz triangular superior.
- A: Matriz conocida.

Así pues, se construyen las siguientes matrices para la versión Doolittle:


<center>  
$
L = 
\begin{pmatrix}
1 & 0 & 0 & 0 & 0 & 0\\
l_{21} & 1 & 0 & 0 & 0 & 0\\
l_{31} & l_{32} & 1 & 0 & 0 & 0\\
l_{41} & l_{42} & l_{43} & 1 & 0 & 0\\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots\\
l_{n1} & l_{n2} & l_{n3} & l_{n4} & ... & 1\\
\end{pmatrix}
$

<center>
$
U = 
\begin{pmatrix}
u_{11} & u_{12} & u_{13} & u_{14} & ... & u_{1n}\\
0 & u_{22} & u_{23} & u_{24} & ... & u_{2n}\\
0 & 0 & u_{33} & u_{34} & ... & u_{3n}\\
0 & 0 & 0 & u_{44} & ... & u_{4n}\\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots\\
0 & 0 & 0 & 0 & ... & u_{nn}\\
\end{pmatrix}
$

De este modo, teniendo en cuenta la ecuación (1), se tiene lo siguiente:

<center>
$
\begin{pmatrix}
1 & 0 & 0 & 0 & 0 & 0\\
l_{21} & 1 & 0 & 0 & 0 & 0\\
l_{31} & l_{32} & 1 & 0 & 0 & 0\\
l_{41} & l_{42} & l_{43} & 1 & 0 & 0\\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots\\
l_{n1} & l_{n2} & l_{n3} & l_{n4} & ... & 1\\
\end{pmatrix}
\ 
\begin{pmatrix}
u_{11} & u_{12} & u_{13} & u_{14} & ... & u_{1n}\\
0 & u_{22} & u_{23} & u_{24} & ... & u_{2n}\\
0 & 0 & u_{33} & u_{34} & ... & u_{3n}\\
0 & 0 & 0 & u_{44} & ... & u_{4n}\\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots\\
0 & 0 & 0 & 0 & ... & u_{nn}\\
\end{pmatrix}
=
\begin{pmatrix}
a_{11} & a_{12} & a_{13} & a_{14} & ... & a_{1n}\\
a_{21} & a_{22} & a_{23} & a_{24} & ... & a_{2n}\\
a_{31} & a_{32} & a_{33} & a_{34} & ... & a_{3n}\\
a_{41} & a_{42} & a_{43} & a_{44} & ... & a_{4n}\\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots\\
a_{n1} & a_{n2} & a_{n3} & a_{n4} & ... & a_{nn}\\
\end{pmatrix}
$

<p style='text-align: justify;'> Los valores de U y L se pueden encontrar operando intercaladamente entre los renglones de U y las columnas de L, de modo que los cálculos se realizan de la siguiente manera, para evitar la acumulación de incógnitas:

1. Primer renglón de U.
2. Primer columna de L.
3. Segundo renglón de U.
4. Segunda columna de L.
5. Tercer renglón de U.
6. Tercer columna de L.

<p style='text-align: justify;'>Así consecutivamente hasta llegar al último renglón de U, y para dichos cálculos se tienen las siguientes ecuaciones...


Para los renglones de U:

<center>
$
u_{ij} = a_{ij}-\sum_{k=1}^{i}\, l_{ik}u_{kj}
$
    
<center>
$
i = 1, 2, 3, ..., n 
$

Para los renglones de L:

<center>
$
l_{ij} = \frac{a_{ij}-\sum_{k=1}^{j}\, l_{ik}u_{kj}}{u_{jj}}
$
    
<center>
$
j = 1, 2, 3, ..., n-1
$

Donde n es el orden de la matriz cuadrada.

### Sustituciones

Ahora tomando en cuenta que una matriz A que multiplica a un vector X, da como resultado un vector B:

<center>
$
[A]\{X\} = \{B\}
$

Al realizar el álgebra correspondiente se pueden llegar a las siguientes expresiones:

<center>
$
[L]\{D\} = \{B\} \rightarrow \ \{D\} \ \ \ \ \ \ [U]\{X\} = \{D\} \rightarrow \{X\}
$

<p style='text-align: justify;'> Donde {X} corresponde a una sustitución hacia atrás, y {D} a una sustitución hacia delante, donde el vector B se define a partir del vector al que esta igualado la matriz A.

<p style='text-align: justify;'>Dichas sustituciones se usarán al final de la obtención de los valores de U y L para poder obtener los valores de X que satisfacen el sistema de ecuaciones

Asi pues, para el calculo de los coeficientes de D y de X se basan en las siguientes ecuaciones

Sustitución hacia adelante:
<center>
$
d_i = b_i - \sum_{k = 1}^{i -1}\,l_{ik}d_{k} ; \ \ \ i = 1, 2, ..., n
$

Sustitución hacia atrás:
<center>
$
x_i = \frac{d_i - \sum_{k = i+1}^{n}\,u_{ik}x_{k}}{u_{ii}}; \ \ \ i = n, n-1, ..., 1
$

```python
import numpy as np

def versionDoolittle(A, B, n):
    sumaU = sumaL = sumaD = sumaX = 0

    U = [[-1, -1, -1],
         [0, -1, -1],
         [0, 0, -1]]

    L = [[1, 0, 0],
         [-1, 1, 0],
         [-1, -1, 1]]

    D = np.full(n, -1.0)
    X =  np.full(n, -1.0)

    for i in range(0, n):
        for j in range(i, n):
            for k in range(0, i):
                sumaU += (L[i][k]*U[k][j])
            U[i][j] = A[i][j] - sumaU
            sumaU = 0

        for j in range(i, n-1):
            for k in range(0, i):
                sumaL += L[j+1][k]*U[k][i]
            L[j+1][i] = (A[j+1][i] - sumaL) / U[i][i]
            sumaL = 0

            
    for i in range(0, n):
        for k in range(0, i):
            sumaD += (L[i][k] * D[k])
        D[i] = (B[i]*1.0) - sumaD
        sumaD = 0

    for i in range(n-1, -1, -1):
        for k in range(i + 1, n):
            u = U[i][k]
            x = X[k]
            sumaX += U[i][k]*X[k]
        X[i] = (D[i] - sumaX) / U[i][i]
        sumaX = 0

    print(f'L: {L}')
    print(f'U: {U}')
    print(f'D: {D}')
    print(f'X: {X}')
    
A = [[10, 1, -5],
    [-20, 3, 20],
    [5, 3, 5]]

B = [1, 2, 6]



versionDoolittle(A, B, len(B))
```

### Sustituciones

Las sustituciones en la versión de Crout siguen el mismo principio que las de Doolittle, pero las ecuaciones para los coeficientes cambian, de modo que se tiene las siguientes fórmulas:

Sustitución hacia adelante:
<center>
$
d_i = \frac{b_i - \sum_{k = 1}^{i-1}\,l_{ik}d_{k}}{l_{ii}}; \ \ \ i = 1, 2, ..., n
$

Sustitución hacia atrás:
<center>
$
x_i = d_i - \sum_{k = i+1}^{n}\,u_{ik}x_{k} ; \ \ \ i = n, n-1, ..., 1
$

```python
import numpy as np

def versionCrout(A, B, n):
    sumaU = sumaL = sumaD = sumaX = 0

    U = [[1, -1, -1],
         [0, 1, -1],
         [0, 0, 1]]

    L = [[-1, 0, 0],
         [-1, -1, 0],
         [-1, -1, -1]]

    D = np.full(n, -1.0)
    X =  np.full(n, -1.0)

    for j in range(0, n):
        for i in range(j, n):
            for k in range(0, j):
                sumaL += L[i][k] * U[k][j]
            L[i][j] = (A[i][j] - sumaL)
            sumaL = 0

        for i in range(j, n-1):
            for k in range(0, j):
                sumaU += (L[j][k] * U[k][i+1])
            U[j][i+1] = (A[j][i+1] - sumaU)/L[j][j]
            sumaU = 0

            
    for i in range(0, n):
        for k in range(0, i):
            sumaD += (L[i][k] * D[k])
        D[i] = (B[i] - sumaD)/L[i][i]
        sumaD = 0

    for i in range(n-1, -1, -1):
        for k in range(i + 1, n):
            sumaX += U[i][k]*X[k]
        X[i] = (D[i] - sumaX)
        sumaX = 0

    print(f'L: {L}')
    print(f'U: {U}')
    print(f'D: {D}')
    print(f'X: {X}')

A = [[3, -6, 7],
    [8, 0, -5],
    [1, -2, 6]]

B = [4, 19, 5]

versionCrout(A, B, len(B))
```

## Método de Krylov y método de las potencias

<p style='text-align: justify;'> Sea un vector $\bar{x}$ que se le aplica una transformación $\bar{x} \rightarrow A\bar{x}$, donde A es una matriz cuadrada, dicha transformación lineal tiene vectores especiales para cada matriz para que sea un procedimiento sencillo, dichos vectores se les conoce como propios.

### Ejemplo

Sea la matriz A = $\begin{pmatrix} 3 & -2 \\ 1 & 0 \end{pmatrix}$ , y los vectores $\bar{u}$ = $\begin{pmatrix} -1 \\ 1 \end{pmatrix}$, $\bar{v}$ = $\begin{pmatrix} 2 \\ 1 \end{pmatrix}$

Las imagenes dadas por las transformación T serían las siguientes:

T($\bar{u}$) = A$\bar{u}$ = $\begin{pmatrix} -5 \\ -1 \end{pmatrix}$

T($\bar{v}$) = A$\bar{v}$ = $\begin{pmatrix} 4 \\ 2 \end{pmatrix}$ = 2$\bar{v}$

De modo que es visible que en $\bar{u}$ hubo una modificación que deformó dicho vector, mientras que en $\bar{v}$ el cambio es una dilatación del vector en un factor de 2, donde dicho factor se le conoce como valor propio. Gráficamente se vería de la siguiente manera:

$$Generado \ por \ Geogebra$$



De este ejemplo se puede extraer lo que es un vector propio ($\bar{v}$) de la matriz cuadrada A, donde se define a un vector propio como:

<center>
$
A\bar{v} = \lambda v \ | \  \bar{v} \neq \bar{0}, \lambda \in \mathbb R
$
   

Tanto el polinomio como los valores y vectores característicos, son datos que nos permiten resolver modelos físicos, en donde el método de Krylov nos otorga los coeficientes de dicho polinomio caracterìstico, diferente a lo que hace el método de las potencias, del cual obtenemos el mayor y el menor de los valores característicos que se obtienen a partir de las raíces del polinomio. De modo que dichos métodos es una alternativa al método analítico visto en Álgebra Lineal.

El objetivo la ecuación antes mencionada es determinar valores de $\lambda$ que resuelvan el sistema con una solución diferente a la trivial, y para la obtención del polinomio caracterísitco, tal como se vió en Álgebra Lineal, se resuelve el siguiente determinante.

<center>
$
|A-\lambda I| = 0 = P(\lambda)
$


Donde el grado del polinomio será igual al grado de la matriz cuadrada, de modo que posee el mismo número de raìces que su grado, y por lo tanto, el mismo número de valores caracterìsticos que su grado, asi pues, se el polinomio tiene la siguiente forma.

<center>
$
P(\lambda) = \sum_{i = 0}^{n}\,{a_{i}\lambda^{n-i}} = 0
$

## Método de Krylov

Este método toma como fundamento el teorema de Cayley-Hamilton, el cual dicta que toda matriz verifica su ecuación caracteristica si al evaluar la matriz en el polinomio este mismo da como resultado 0.

<center>
P(A) = 0

Asi pues, el siguiente polinomio caracterìstico resultado de la matriz A:

<center>
$
P(\lambda) = \lambda^{n} + b_{1}\lambda^{n-1} + b_{2}\lambda^{n-2} + ... + b_{n-1}\lambda + b_{n} = 0
$

Aplicando el teorema de Cayley Hamilton usando la matriz A, se obtiene el siguiente polinomio.

<center>
$
P(A) = A^{n} + b_{1}A^{n-1} + b_{2}A^{n-2} + ... + b_{n-1} + b_{n}I = 0
$

Donde los coeficientes de b son desconocidos, formando un sistema de ecuaciones lineales.

En cuanto la polinomio P(A), únicamente con el propósito de simplificar los cálculos, el polinomio se multiplica por un vector $\bar{y}$, compatible con el orden de la matriz A y diferente al vector $\bar{0}$

<center>
$
\bar{y}P(A) = A^{n}\bar{y} + b_{1}A^{n-1}\bar{y} + b_{2}A^{n-2}\bar{y} + ... + b_{n-1}\bar{y} + b_{n}I\bar{y} = 0
$

Donde si bien el vector $\bar{y}$ puede ser definido arbitrariamente, es recomendable usar las siguientes propuestas

<center>
$
\bar{y} = \begin{pmatrix} 1\\0\\0\\0\\ \vdots \\ 0 \end{pmatrix} \ \ \  \bar{y} = \begin{pmatrix} 1\\1\\1\\1\\ \vdots \\ 1 \end{pmatrix}
$
    

De modo que los términos de $\bar{y}P(A)$ resultan en vectores, donde los mismos se pueden obtener a partir de la siguiente ecuación de recursividad.

<center>
$
A^k\bar{y} = A[A^{k-1}\bar{y}]
$

Asi pues, $A^{k-1}$ es un vector de $n\times1$, donde al sumar los vectores de la ecuación, se obtiene un sistema de n ecuaciones, donde al resolverlo se obtienen los coeficientes de la ecuación característica

## Método de las potencias

El método de las potencias es un método iterativo para obtener los valores caraterísticos de una matriz cuadrada, una buena ventaja que tienen este tipo de métodos es que además de se obtiene el vector correspondiente asociado a dicho valor característico.

Al igual que en Krylov, este método usa la ecuación $A\bar{v} = \lambda\bar{v}$, pero a diferencia de Krylov, aqui se toma un vector inicial $\bar{v}^{(0)} \ \neq  \ \bar{0}$ de la siguiente forma

<center>
$\bar{v}^{(0)} = \begin{pmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{pmatrix}$

Al igual que en el método de Krylov, los coeficientes del vector $\bar{v}$ se definen arbitrariamente, pero se recomiendan usar los siguientes vectores

<center>
$
\bar{v} = \begin{pmatrix} 1\\0\\0\\0\\ \vdots \\ 0 \end{pmatrix} \ \ \  \bar{v} = \begin{pmatrix} 1\\1\\1\\1\\ \vdots \\ 1 \end{pmatrix}
$

Por lo que al hacer la sustitución en la ecuación que define a un valor propio, se puede obtener la aproximación siguiente:

<center>
$
A\bar{v}^{(0)} = \lambda^{(1)}\bar{v}^{(1)} = \lambda^{(1)} \begin{pmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{pmatrix}
$


Y teniendo el vector del producto realizado entre $\lambda$ y $\bar{v}$, se selecciona el mayor elemento, en valor absoluto, de dicho vector resultando, para asignarlo a la aproximación de $\lambda$, donde la aproximación se hace con el signo que naturalmente acompaña al número seleccionado, para posteriormente normalizar el vector obtenido (dividirlo entre la aproximación escogida de $\lambda$), obteniendo asi la aproximación correspondiente al vector $\bar{v}$.

### Mayor Valor

Asi pues, dado que este método se define como uno de aproximaciones sucesivas, la ecuación que lo rige es la siguiente:

<center>
$
A\bar{v}^{(k-1)} = \lambda^{(k)}\bar{v}^{(k)},  \ \ k = 1, 2, 3, ...
$

### Menor valor

En cuanto a la obtención del menor valor se refiere, se basa en los mismos principio que la obtención del menor valor, el único cambio relevante se da en la ecuación de recurrencia, que se define como:

<center>
$
A^{-1}\bar{v}^{(k-1)} = (\frac{1}{\lambda})^{(k)}\bar{v}^{(k)},  \ \ k = 1, 2, 3, ...
$

Donde en ambos casos, $\lambda$ resulta en el mayor valor hasta que el error absoluto entre dos aproximaciones sucesivas sea menor que una tolerancia preestablecida.

### Interpolación 
La interpolación polinomial es el método más común para estimar valores intermedios entre datos definidos por puntos. Recordando que la fórmula general para un polinomio de n-ésimo grado es $$f(x)=a_0+a_1x+a_2x^2+...+a_nx^n$$
Dados n + 1 puntos, hay un solo polinomio de grado n que pasa a través de todos los puntos. La interpolación polinomial consiste en determinar el polinomio único de n-ésimo grado que se ajuste a n+1 puntos. Este polinomio, entonces, proporciona una fórmula para calcular valores intermedios. 

## Interpolación con incrementos variables (polinomio de Lagrange).
Se analizarán dos alternativas que son muy adecuadas para implementarse en computadora: los polinomios de Lagrange y los polinomios interpolantes.  

Sí se presenta una función tabulada en forma   

| x                | y = F(x) |
| --- | --- |
| $x_0$ |$y_0$ | 
| $x_1 = x_0+h_0$ | $y_1$ |
| $x_2=x_1+h_1$ | $y_2$ |   
| $\vdots$ | $\vdots$ |
| $x_n=x_{n-1}+h_{n-1}$ | $y_n$ |  
 
Donde  
$h_0=x_1 - x_0 $  
$h_1=x_2 - x_1 $  
$\vdots$  
$h_{n-1}=x_n - x_{n-1} $  
y en la que no necesariamente se cumple que   
$h_0 = h_1 = h_2 = \dots = x_{n-1} $   
Entonces el polinomio n que pasa por todos los n+1 puntos es: $$y =b_0x^n+b_1x^{n-1}+b_2x^{n-2}+\dots+b_{n-1}+b_n$$  
O bien: 
$y = a_0(x-x_1)(x-x_2)(x-x_3)\dots(x-x_n)$  
$ + a_1(x-x_0)(x-x_2)(x-x_3)\dots(x-x_n)$  
$ + a_2(x-x_0)(x-x_1)(x-x_3)\dots(x-x_n)$  
        $ +       \vdots       $  
$ + a_n(x-x_0)(x-x_1)(x-x_2)\dots(x-x_{n-1})$

Los coeficientes $a_0,a_1,\dots,a_n$ se determinan de tal forma que el polinomio pase por todos y cada uno de los puntos conocidos de la función; entonces, sí se evalúa la ecuación anterior para $x = x_0$ se obtiene: $$y = a_0(x_0-x_1)(x_0-x_2)(x_0-x_3)\dots(x_0-x_n)$$ Se despeja $a_0$: $$a_0 = \frac{y_0}{(x_0-x_1)(x_0-x_2)(x_0-x_3)\dots(x_0-x_n)}$$ Ahora se evalúa $x = x_1$ para obtener $y_1$: $$y_1 = a_1(x_1-x_0)(x_1-x_2)(x_1-x_3)\dots(x_1-x_n)$$ Por lo tanto: $$a_1 = \frac{y_1}{(x_1-x_0)(x_1-x_2)(x_1-x_3)\dots(x_1-x_n)}$$ En forma análoga se obtienen los demás coeficientes de la ecuación A; el último de ellos se obtiene al evaluar esta expresión para $x = x_n$:  $$a_n = \frac{y_n}{(x_n-x_0)(x_n-x_1)(x_n-x_2)\dots(x_n-x_{n-1})}$$ Sustituyendo los valores de $a_0,a_1,a_2,\dots,a_n$ en la ecuación A, se obtiene: <span style="color:blue">$$y=  \sum_{i=0}^{n} \left\{\prod_{j=0 j≠ i}^{n} \frac{x-x_j}{x_i-x_j}\right\}y_i$$</span> 


### Código para encontrar valores determinados por polinomio de Lagrange
**<span style="color:green">Ejemplo:</span>** Dada la función definida por la siguiente tabla, encontrar el valor de la función para x=3.    

| x     | y = f(x) |
| --- | --- |
| 0 | 5 |
| 1 | 7 |
| 2 | 9 |
| 5 | 15 |

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
xi=np.array([0,1,2,5])
fi=np.array([5,7,9,15])
n=len(xi)
x=sym.Symbol('x')
polinomio=0
for i in range(0,n):
    N=1 #numerador
    D=1 #denominador
    for j in range(0,n):
        if (i!=j):
            N=N*(x-xi[j])
            D=D*(xi[i]-xi[j])
        F=(N/D)*fi[i] #fracción
    polinomio=polinomio+F
polisimple=sym.expand(polinomio) # para la simplificación del polinomio

# Impresión de la solución
print("Polinomio de Lagrange")
print(polinomio)
print("Polinomio de Lagrange reducido:")
print(polisimple)

# Graficar la solución
Px=sym.lambdify(x,polinomio) #permite evaluar a x dentro del polinomio obtenido
muestras=25
xl=np.min(xi) #límite inferior de la gráfica para x
xu=np.max(xi) #límite superior de la gráfica para x
Pol_xi=np.linspace(xl,xu,muestras) # se genera 50 divisiones entre el límite inferior y superior p evaluar el pol.
Pol_yi=Px(Pol_xi) #se evalua el polinomio en el dominio generado en Pol_xi

plt.plot(xi,fi,'o') #se grafican los puntos de la función tabular
plt.plot(Pol_xi,Pol_yi)
plt.show()
```

## Integración numérica: Fórmulas de integración trapecial y de Simpson

Es posible operar con una función tabular sin la necesidad de obtener su función analítica, para esto se presentarán las herramientas para integrar dicha función; estas requieren de las consideraciones básicas del cálculo integral y por tratarse de técnicas numéricas su resultado es el valor del área bajo la curva de la función; en consecuencia, es indispensable contar con el intervalo de integración respectivo.   
Para obtener un valor aproximado de la integral de $F(x)$ en el intervalo $x_0 ≤ x ≤ x_n$ se parte de la fórmula de interpolación de Newton: $$f(x) = y_0 + \binom{k}{1} Δy_0 + \binom{k}{2} Δ^2y_0 + \binom{k}{3} Δ^3y_0 + \dots + \binom{k}{j} Δ^jy_0 +  $$  
Integrando y sustituyendo lo anterior queda: $$\int_{x_0}^{x_1} f(x)dx = \frac{h}{2} [y_0 + y_1] + e_r$$  
La ecuación anterior es igual al área bajo la curva entre los puntos de coordenadas $(x_0, y_0)$ y $(x_1, y_1)$; ambos unidos por una recta definida por la ecuación $$y = y_0 + kΔy_0$$  
Su interpretación geometrica es:  

$$Generado \ por \ Paint \ y \ Power \ Point$$

Al integrar y simplificar entre $x_1$ y $x_2$ se obtiene: $$\int_{x_0}^{x_n} F(x)dx= \frac{h}{2}\left\{ y_0 + y_n + 2 \sum_{i=1}^{n-1} y_i\right\}+e_r$$  
donde $e_r$, es la suma algebraica de los errores en cada integración parcial. La ecuación anterior se conoce como fórmula de integración trapecial.  
Para obener una mejor aproximación al valor de la integral definida se deben considerar las segundas diferencias $$\int_{x_0}^{x_n} f(x)dx = h [ ny_0 + \frac{n^2}{2} Δy_0 + (\frac{n^3}{6} - \frac{n^2}{4}) Δ^2y_0 ] + e_r$$   
Sustituyendo y simplificando queda:  $$\int_{x_0}^{x_2} F(x) dx = \frac{h}{3} (y_0 +4y_1 + y_2)$$  
Su interpretación geometrica es:  





$$Generado \ por \ Paint \ y \ PowerPoint$$


Al integrar y simplificar entre $x_2$ y $x_4$ se obtiene: $$\int_{x_0}^{x_n} F(x)dx= \frac{h}{3}\left\{ y_0 + y_n + 4 \sum ordenadas con indice impar + 2\sum ordenadas con indice par \right\}+e_r$$  donde $e_r$, es la suma algebraica de los errores en cada integración parcial. Esta ecuación es conocida como fórmula de integración de Simpson 1/3.  

A diferencia de la fórmula de integración trapecial, la fórmula Simpson 1/3 es aplicable para valores pares de $n$, ya que de lo contrario no se podrá obtener el valor de la última integral.  

Si se considera hasta las terceras diferencias en la ecuación y se integra entre $x_0$ y $x_3$ se tiene: $$\int_{x_0}^{x_3}F(x)dx = h [3y_0 + \frac{9}{2} Δy_0 + (\frac{27}{6} - \frac{9}{2}) Δ^2y_0+ (\frac{81}{24} - \frac{27}{6} + \frac{9}{2})Δ^3y_0 ] + e_r$$   
Al sustituir y simplificar queda $$\int_{x_0}^{x_3}F(x)dx = \frac{3}{8}h[y_0 + 3y_1 + 3y_2 + y_3] + e_r$$
Su interpretación geométrica es:

$$Generado \ por \ Paint \ y \ Power \ Point$$

Al sustituir los valores de las integrales y sustituir queda: 
$$\int_{x_0}^{x_n} F(x)dx= \frac{3}{8}h\left\{ y_0 + y_n + 2 \sum ordenadas con indice multiplo de 3 + 3\sum resto de ordenadas \right\}+e_r$$  donde $e_r$ es la suma algebraica de los errores en cada integración parcial. Esta expresión se conoce como fórmula de integración Simpson 3/8 y es aplicable para valores de $n$ múltiplos de tres.

### Código para resolver integrales con Simpson 1/3

```python
#Simpson 1/3

#Define la funcion a integrar 
def f(x):
    return 1/(1 + x**2)

#Implementa Simpson 1/3 
def simpson13(x0,xn,n):
    #calculando espaciamiento
    h = (xn - x0) / n
    
    #Suma
    integration = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + i*h
        
        if i%2 == 0:
            integration = integration + 2 * f(k)
        else:
            integration = integration + 4 * f(k)
    
    #valor de la integral
    integration = integration * h/3
    
    return integration
    
# Input section
lower_limit = float(input("Ingresa el limite inferior de la integral -> "))
upper_limit = float(input("Ingresa el limite superior de la integral -> "))
sub_interval = int(input("Ingresa el número se subintervalos: "))

result = simpson13(lower_limit, upper_limit, sub_interval)
print("Integrando por Simpson 1/3 se obtiene: %0.6f" % (result) )
```

### Código para resolver integrales con Simpson 3/8

```python
#Simpson 3/8

#Define la funcion a integrar
def f(x):
    return 1/(1 + x**2)

# Implementar Simpson 3/8
def simpson38(x0,xn,n):
    #Tamaño de espaciamiento
    h = (xn - x0) / n
    
    #Suma
    integration = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + i*h
        
        if i%3 == 0:
            integration = integration + 2 * f(k)
        else:
            integration = integration + 3 * f(k)
    
    #Valor de la integral
    integration = integration * 3 * h / 8
    
    return integration
    
# Input section
lower_limit = float(input("Ingresa el limite inferior de la integral -> "))
upper_limit = float(input("Ingresa el limite superior de la integral -> "))
sub_interval = int(input("Ingresa el número se subintervalos: "))

result = simpson38(lower_limit, upper_limit, sub_interval)
print("Integrando por Simpson 3/8 se obtiene: %0.6f" % (result) )
```

### Código para resolver integrales con fórmula trapecial

```python
#Metodo Trapecial

#Define la funcion a integrar
def f(x):
    return 1/(1 + x**2)

#Implementa la formula trapecial
def trapezoidal(x0,xn,n):
    #Espaciamiento
    h = (xn - x0) / n
    
    #Suma
    integration = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + i*h
        integration = integration + 2 * f(k)
    
    #Valor de la integral
    integration = integration * h/2
    
    return integration
    
# Input section
lower_limit = float(input("Ingresa el limite inferior de la integral -> "))
upper_limit = float(input("Ingresa el limite superior de la integral -> "))
sub_interval = int(input("Ingresa el número se subintervalos: "))

result = trapezoidal(lower_limit, upper_limit, sub_interval)
print("Integrando por el método trapecial se obtiene: %0.6f" % (result) )
```

## <center> Conceptos Básicos

### Ecuación diferencial

Es aquella que relacioa dos o más variables en términos de derivadas o diferenciales, donde se puede categorizar según sus propiedades:

- Ordinaria: Solo hay una variable independiente; por lo que sus derivadas serán totales.

- Parcial: Existen dos o más variables independientes; sus derivadas serán parciales.



Del mismo modo, es posible diferenciar diferentes características de una ecuación diferencial:

- Orden: Es el número de derivadas más alta que se encuentra en uno solo de los términos de la ecuación.

- Grado: Es el exponente al cual esta elevado el término de mayor orden de derivación en la ecuación (no todas las ecuaciones diferenciales poseen un grado).

En cuanto a su clasificación, tambien es posible determinar su linealidad, de modo que esta se mantiene siempre y cuando no aparezcan potencias de la variable dependiente y sus derivadas, al igual que tampoco hay productos de la variable dependiente por sus derivadas o productos entre derivadas.

###  Solución de una ecuación diferencial

Una ecuación diferencial tiene como solución a una función que representa a una familia de funciones, la cual se le denomina solución general, que esta excenta de operadores diferenciales o integrales, además de verificar a la misma ecuación diferencial.

Sea la ecuación diferencial:

$y'(x) - sen(x) = 0$

Su solución general es:

$y(x) = -cos(x) + C$

Y dicha solución representa a la siguiente familia de curvas:

$$Generado \ por \ Geogebra$$

$y(x) = -cos(x) + C \ ; \ y(x_0) = y_0 \$

$$Generado \ por \ Geogebra$$

Para determinar una solución particular, la cual es una curva de la solución general, es necesario determinar una condición inicial, con lo cual será posible determinar el valor de $C$.

## Polinomio de Taylor como solución de una ecuación diferencial

El polinomio de Taylor se puede usar para aproximar a todas aquellas ecuaciones diferenciales con condiciones iniciales y que tengan una solución $y(x)$ tal que verifique idénticamente la ecuación diferencial, donde el polinomio tiene la siguiente forma:

$
y(x) = T_{n} = [y(x;x_{0})] \\
$
$
[y(x;x_{0})] = y(x_{0}) + y'(x_{0})(x - x_{0}) + \frac{y''(x_{0})(x-x_{0})^2}{2!} + \frac{y'''(x_{0})(x-x_{0})^3}{3!} + ... + + \frac{y^{n}(x_{0})(x-x_{0})^n}{n!}
$

Donde $T_{n}$ es correcta siempre y cuando se evalúe en $x = x_{0}$, pero a medida el polinomio se aleje de tal punto, el error de la aproximación aumenta.

Para  $T_{n}$, las incognitas son los valores de $y(x_0)$, $y'(x_0)$, ..., $y^n(x_0)$, las cuales se pueden obtener con las condiciones iniciales, al igual que despejando la derivada de mayor orden y sustituyendo las condiciones iniciales, donde despuès se deriva y aplicando el paso anterior para los coeficientes de $y^{(i)}$ 

```python
import numpy as np
import math
import matplotlib.pyplot as plt

def P_T(x):
    return (7/4)*x**3-(3/8)*x**2+(9/16)*x+(19/32)
h=0.1
n=7

x=np.zeros(n)
y=np.zeros(n)

x[0]=0.5
y[0]=1.0
print('{:^12} {:^10}'.format(' x ', 'y_aprox'))
print('{:10.3f} {:10.5f}'.format(x[0], y[0]))
for i in range(1,n):
    x[i]=x[i-1]+h
    y[i]=P_T(x[i])
    print('{:10.3f} {:10.5f}'.format(x[i], y[i]))
plt.scatter(x,y, color='black') #puntos para la solución aproximada
plt.plot(x,y, "red") #línea continua para la solución aproximada
```

## Método de Euler

Estos métodos sirven para dar solución a Ecuaciones Diferenciales de Primer Orden, donde también se toman en cuenta a aquellas de orden n si se toman como un sistema de $n$ ecuaciones diferenciales de primer orden , donde una EDO se puede representar de la siguiente manera:


<center>
$
y' = f(x, y)
$
    
Donde toda EDO se puede llevar a esa forma con manipulacione algebraìcas.

El método de euler es uno de los más sencillos para resolver ecuaciones diferenciales de primer orden con condiciones iniciales, donde se integra a la ecuación diferencial entre ($x_i$) y ($x_{i+1}$):

$$
\int_{x_i}^{x_{i+1}} y'(x)\,dx = \int_{x_i}^{x_{i+1}} f(x_i, y_i)\,dx
\\
y_{i+1} - y_i = \int_{x_i}^{x_{i+1}} f(x_i, y_i)\,dx
$$

Donde 

$$
\int_{x_i}^{x_{i+1}} f(x_i, y_i)\,dx = h[f(x_i, y_i) + \frac{1}{2}\{f(x_{i + 1}, y_{i+1})- f(x_{i}, y_{i}) + ...]
$$


Tomando en cuenta unicamente considerando unicamente el primer sumando de la expresión anterior, llegamos a lo que es el método de Euler
$$
\\
\int_{x_i}^{x_{i+1}} f(x_i, y_i)\,dx = hf(x_i, y_i) + e_r
$$

Por lo que al final se obtiene la siguiente ecuación, la cual es propia del método de Euler (se desprecia el error).

<center>
$
y_{i+1} = hf(x_i, y_i) + y_i \ \ ; \ \ i = 0, 1, 2, ...
$

```python
import numpy as np
import math
import matplotlib.pyplot as plt

def f(x,y):
  return x+2*x*y
def S(x):
  return 1.1682*math.exp(x**2)-0.5

n=6
h=0.1

x=np.zeros(n)
y=np.zeros(n)
yS=np.zeros(n)


x[0]=0.5
y[0]=1
yS[0]=1

print('{:^12} {:^10} {:^10}'.format(' x ', 'y_aprox', 'y_analitica'))
print('{:10.3f} {:10.5f} {:10.5f}'.format(x[0], y[0], yS[0]))
for i in range(1,n):
  x[i]=x[i-1]+h
  y[i]=y[i-1]+h*f(x[i-1],y[i-1])
  yS[i]=S(x[i])
  print('{:10.3f} {:10.5f} {:10.5f}'.format(x[i], y[i], yS[i]))
plt.scatter(x,y)
plt.plot(x,yS, color='red')
```

## Método de Euler Modificado

En el método de Euler modificado, al resolver la integral de $f(x_i, y_i)$, se usa la fórmula de integración trapecial, donde al sustituir y simplificar se obtiene lo siguiente (despreciando la suma del error):

<center>
$
y_{i+1} = \frac{h}{2}[f(x_i, y_i) + f(x_{i+1}, y_{i+1})] + y_i \ \ ; \ \ i = 0, 1, 2, ...
$

Pero para su correcta utilización, es necesario el uso en un primer paso del método de euler modificado debido a que la ecuación presentada depende de si misma para su solución, de modo que el método de Euler se puede ver como una primera aproximación, mientras que el método modificado es una corrección a esa primera aproximación para poder disminuir el error numérico.

<center>
$
y_{i+1_{p}} = hf(x_i, y_i) + y_i
$  


<center>
$
y_{i+1_{c}} = \frac{h}{2}[f(x_i, y_i) + f(x_{i+1}, y_{i+1})] + y_i
$

<center>
$
i = 0, 1, 2, ...
$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x,y):
  return x+2*x*y
def S(x):
  return 1.1682*math.exp(x**2)-0.5

n=51
h=0.01

x=np.zeros(n)
yp=np.zeros(n)
yc=np.zeros(n)
yS=np.zeros(n)

x[0]=0.5
yc[0]=1.0
yS[0]=1.0

print('{:^12} {:^10} {:^10}'.format(' x ', 'y_numerica', 'y_analitica'))
print('{:10.3f} {:10.5f} {:10.5f}'.format(x[0], yc[0], yS[0]))
for i in range(1,n):
  x[i]=x[i-1]+h
  yp[i]=yc[i-1]+h*f(x[i-1],yc[i-1])
  yc[i]=yc[i-1]+(h/2)*(f(x[i-1],yc[i-1])+f(x[i],yp[i]))
  yS[i]=S(x[i])
  print('{:10.3f} {:10.5f} {:10.5f}'.format(x[i], yc[i], yS[i]))
plt.scatter(x,yc)
plt.plot(x,yS)
```
